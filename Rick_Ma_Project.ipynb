#### Underdamped System

```python
#Underdamped System
%matplotlib inline
from math import *
import numpy as np
import matplotlib.pyplot as plt

gamma = 0.7
k = 48.

#RK4 Constants
c1 = 1./6
c2 = 2./6
c3 = 2./6
c4 = 1./6
a2 = 1./2
a3 = 1./2
a4 = 1.
b21 = 1./2
b31 = 0.
b32 = 1./2
b41 = 0.
b42 = 0.
b43 = 1.


F_over_m = lambda t: 0                       #5*sin(2*pi*sqrt(k)*t)
fxy_x = lambda t, x, v: v                         #this equation for dx/dt = v, t = independent, x = primary dependent, v = dependent
fxy_v = lambda t, x, v: F_over_m(t)-2*gamma*v - k*x        #this is equation for dv/dt = -2*gamma*v - k*t
                                                      # t = independent primary dependent = v, x is also dependent
def rk4_x(ti, xi, vi, h):
    K1 = fxy_x(ti, xi, vi)
    K2 = fxy_x(ti+a2*h, ti+b21*K1*h, vi)
    K3 = fxy_x(ti+a3*h, xi+b31*K1*h+b32*K2*h, vi)
    K4 = fxy_x(ti+a4*h,xi+b41*K1*h+b42*K2*h+b43*K3*h, vi)   
    xip1 = xi+(c1*K1+c2*K2+c3*K3+c4*K4)*h
    return xip1


def rk4_v(ti, xi, vi, h):
    K1 = fxy_v(ti, xi, vi)
    K2 = fxy_v(ti+a2*h, xi, vi+b21*K1*h)
    K3 = fxy_v(ti+a3*h, xi, vi+b31*K1*h+b32*K2*h)
    K4 = fxy_v(ti+a4*h, xi, vi+b41*K1*h+b42*K2*h+b43*K3*h)   
    vip1 = xi+(c1*K1 + c2*K2 + c3*K3 + c4*K4)*h
    return vip1

a = 0.0    #initial value for x
b = 2.1    #final value for x 
n = 5000
h = (b-a)/(n-1)
ti = 0.0
tf = 5.0
dt = (tf - ti)/n
t = np.zeros(n+1)
x = np.zeros(n+1)
v = np.zeros(n+1)
x_iv = 0.0
v_iv = 0.5
t[0] = a
x[0] = x_iv
v[0] = v_iv


for i in range(1, n+1):
    t[i] = a + i*dt
    x[i] = rk4_x(t[i-1], x[i-1], v[i-1], h)
    v[i] = rk4_v(t[i-1], x[i-1], v[i-1], h)

fig = plt.figure(figsize = (6,6))
plt.title("Underdamped System (Displacement)")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.plot(t, x, label = "x vs. t", color = "peru", linewidth = "5.0")
plt.legend()
plt.tight_layout()
plt.show()

fig = plt.figure(figsize = (6,6))
plt.title("Underdamped System (Velocity)")
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.plot(t, v, label = "v vs. t", color = "lime", linewidth = "5.0")
plt.legend()
plt.tight_layout()
plt.show()
```
